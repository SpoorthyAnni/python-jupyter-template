### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt
import numpy as np

data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]

# Error handling: Check if data is a list of numbers
if not isinstance(data, list):
    raise TypeError("data must be a list")
if not all(isinstance(i, (int, float)) for i in data):
    raise TypeError("all elements of data must be numbers")

# Check for negative values (optional)
if any(i < 0 for i in data):
    raise ValueError("Negative values are not allowed in this dataset")

plt.figure(figsize=(7, 5))
bins = np.arange(min(data), max(data) + 1.5)  # Creating bin edges for each integer

counts, _, patches = plt.hist(data, bins=bins, color='cornflowerblue', edgecolor='black', alpha=0.8)

# Annotate bars with frequency counts
for patch, count in zip(patches, counts):
    height = patch.get_height()
    if height > 0:  # Avoid displaying labels on empty bins
        plt.text(patch.get_x() + patch.get_width() / 2, height, f'{int(count)}', 
                 ha='center', va='bottom', fontsize=12, color='black')

plt.title('Histogram of Data', fontsize=14, fontweight='bold')
plt.xlabel('Values', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(range(min(data), max(data) + 1))
plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```
