### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]
categories = ['A', 'B', 'C', 'D']
values = [5, 3, 9, 6]
labels = ['Apple', 'Banana', 'Cherry', 'Date']
sizes = [10, 15, 7, 5]
colors = ['blue', 'red', 'green', 'orange']

# Error handling: Check data validity
if len(x) != len(y):
    raise ValueError("x and y must have the same length for the line plot")
if len(categories) != len(values):
    raise ValueError("categories and values must have the same length for the bar plot")
if len(labels) != len(sizes):
    raise ValueError("labels and sizes must have the same length for the pie chart")

# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 10))

# Line plot
axs[0, 0].plot(x, y, marker='o', linestyle='-', color='blue')
axs[0, 0].set_title('Line Plot', fontsize=14)
axs[0, 0].set_xlabel('X Values')
axs[0, 0].set_ylabel('Y Values')
axs[0, 0].grid(True, linestyle='--', alpha=0.6)

# Bar plot
axs[0, 1].bar(categories, values, color=colors)
axs[0, 1].set_title('Bar Plot', fontsize=14)
axs[0, 1].set_xlabel('Categories')
axs[0, 1].set_ylabel('Values')

# Scatter plot
axs[1, 0].scatter(x, y, color='red', marker='s')
axs[1, 0].set_title('Scatter Plot', fontsize=14)
axs[1, 0].set_xlabel('X Values')
axs[1, 0].set_ylabel('Y Values')
axs[1, 0].grid(True, linestyle='--', alpha=0.6)

# Pie chart
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, wedgeprops={'edgecolor': 'black'})
axs[1, 1].set_title('Pie Chart', fontsize=14)

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```
